```python
from scipy.stats import ttest_ind

# 'df' is DataFrame
threshold = 1000
margin = 100

# Create two groups: above_threshold and below_threshold
above_threshold = df[df['expected_recovery_amount'] >= threshold + margin]['age']
below_threshold = df[df['expected_recovery_amount'] < threshold - margin]['age']

# Perform independent two-sample t-test
t_statistic, p_value = ttest_ind(above_threshold, below_threshold, equal_var=False)

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")
```

# conclusion - The p-value is lower than T-statistic value, so there is large significant difference in the distribution of sexes above and below the threshold.

# Statistical test: sex vs. expected recovery amount

```python
from scipy.stats import chi2_contingency

# 'df' is DataFrame
cross_tab = pd.crosstab(df['sex'], df['recovery_strategy'])

print(cross_tab)

# Perform chi-square test
chi2_stat, p_value, _, _ = chi2_contingency(cross_tab)

print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")
```

# conclusion - The p-value is lower than Chi-square value, so there is large significant difference in the distribution of sexes above and below the threshold.

# Regression modeling: no threshold

```python
import statsmodels.api as sm

#  'df' is  DataFrame
X = sm.add_constant(df['expected_recovery_amount'])  # Adding a constant term to the predictor
y = df['actual_recovery_amount']

# Build linear regression model
model = sm.OLS(y, X).fit()

# Predict using the model
predictions = model.predict(X)

# Print out the model summary statistics
print(model.summary())
```

# Regression modeling: adding true threshold

```python
# 'df' is DataFrame
df['indicator_1000'] = (df['expected_recovery_amount'] >= 1000).astype(int)

# Build linear regression model
X = df[['expected_recovery_amount', 'indicator_1000']]
X = sm.add_constant(X)  # Adding a constant term to the predictor
y = df['actual_recovery_amount']

threshold_model = sm.OLS(y, X).fit()

# Confidence intervals
confidence_intervals = threshold_model.conf_int(alpha=0.05)

# Print the model summary
print(threshold_model.summary())
print(f"\nMean Confidence Intervals:\n{confidence_intervals}")
```
