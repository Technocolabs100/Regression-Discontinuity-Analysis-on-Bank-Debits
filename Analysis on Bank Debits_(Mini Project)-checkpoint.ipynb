```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("bank_data.csv")
```

# Graphical exploratory data analysis

```python
# Filter data for the specified range
selected_range = (df['expected_recovery_amount'] >= 0) & (df['expected_recovery_amount'] <= 2000)
data_subset = df[selected_range]
```

```python
# Create scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='expected_recovery_amount', y='age', data=data_subset, hue='recovery_strategy', palette='viridis')
plt.axvline(x=1000, color='red', linestyle='--', linewidth=2, label='Threshold ($1000)')
plt.title('Scatter Plot of Age vs Expected Recovery Amount')
plt.xlabel('Expected Recovery Amount')
plt.ylabel('Age')
plt.legend()
plt.show()
```

# Statistical test: age vs. expected recovery amount

```python
from scipy.stats import kruskal

# Define the ranges
above_threshold_range = df[(df['expected_recovery_amount'] > 1000) & (df['expected_recovery_amount'] <= 1100)]['age']
below_threshold_range = df[(df['expected_recovery_amount'] >= 900) & (df['expected_recovery_amount'] < 1000)]['age']

# Perform the Kruskal-Wallis test
statistic, p_value = kruskal(above_threshold_range, below_threshold_range)

# Print the results
print(f"Kruskal-Wallis Statistic: {statistic}")
print(f"P-value: {p_value}")
```

 Since the p-value (0.0755) is greater than the conventional significance level of 0.05, we do not have enough evidence to reject the null hypothesis. This suggests that, based on the age variable, there is not a significant difference between the groups just above and just below the 1000 dollar threshold.

# Statistical test: sex vs. expected recovery amount

```python
from scipy.stats import chi2_contingency

# Assuming df is your DataFrame

# Create a categorical variable for the range (900 to 1100)
df['within_range'] = pd.cut(df['expected_recovery_amount'], bins=[900, 1100], labels=['Within Range'])

# Create a contingency table
contingency_table = pd.crosstab(df['sex'], df['within_range'])

# Perform the chi-squared test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-Squared Statistic: {chi2}")
print(f"P-value: {p}")

# Interpret the results
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis. There is a significant difference in gender distribution within the specified range.")
else:
    print("Fail to reject the null hypothesis. No significant difference in gender distribution within the specified range.")
```

The p-value of 1.0 further supports this, indicating that there is no evidence to reject the null hypothesis that gender distribution is the same within the specified range.

In practical terms, this suggests that there is no apparent jump or discontinuity in the percentage of male customers compared to female customers within the analyzed range. The gender distribution appears to be consistent across the range from 900 dollar to 1100 dollar.

# Exploratory graphical analysis: recovery amount

```python
# Filter data for the specified range
scatter_data = df[(df['expected_recovery_amount'] >= 900) & (df['expected_recovery_amount'] <= 1100)]

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='expected_recovery_amount', y='actual_recovery_amount', data=scatter_data, hue='recovery_strategy', palette='viridis')
plt.title('Scatter Plot of Expected vs. Actual Recovery Amount')
plt.xlabel('Expected Recovery Amount')
plt.ylabel('Actual Recovery Amount')
plt.axvline(x=1000, color='red', linestyle='--', label='Threshold')
plt.legend()
plt.show()
```

Yes, there appears to be a discontinuity (jump) around the 1000 dollar threshold in the scatter plot of Expected Recovery Amount (Y) vs. Actual Recovery Amount (X) for Expected Recovery Amounts between 900 dollar to 1100 dollar.

The data points for expected recovery amounts below 1000 dollar are clustered tightly together, while the data points for expected recovery amounts above 1000 dollar are more spread out. This suggests that there is a difference in the distribution of actual recovery amounts for customers below and above the threshold.

```python
from scipy.stats import kruskal

# Define the two windows
window1 = (df['expected_recovery_amount'] >= 900) & (df['expected_recovery_amount'] <= 1100)
window2 = (df['expected_recovery_amount'] >= 950) & (df['expected_recovery_amount'] <= 1050)

# Perform Kruskal-Wallis test for each window
statistic1, p_value1 = kruskal(df[window1]['actual_recovery_amount'], df[~window1]['actual_recovery_amount'])
statistic2, p_value2 = kruskal(df[window2]['actual_recovery_amount'], df[~window2]['actual_recovery_amount'])

# Print the results
print(f"Window 1 - Kruskal-Wallis Statistic: {statistic1}, p-value: {p_value1}")
print(f"Window 2 - Kruskal-Wallis Statistic: {statistic2}, p-value: {p_value2}")
```

The Kruskal-Wallis test results indicate statistically significant differences in the distribution of actual recovery amounts between the specified windows and the rest of the data. The p-values are very close to zero, suggesting strong evidence against the null hypothesis that the distributions are the same.

# Regression modeling: no threshold

```python
import statsmodels.api as sm
import pandas as pd

# Define the input (independent variable) and output (dependent variable)
X = df[['expected_recovery_amount']]
y = df['actual_recovery_amount']

# Add a constant term to the independent variable
X = sm.add_constant(X)

# Fit the OLS regression model
ols_model = sm.OLS(y, X).fit()

# Print the summary
print(ols_model.summary())
```

# Regression modeling: adding true threshold

```python
# Add an indicator variable for being above the threshold
df['above_threshold'] = (df['expected_recovery_amount'] > 1000).astype(int)

# Model with a threshold indicator
model_with_threshold = sm.OLS(df['actual_recovery_amount'], sm.add_constant(df[['expected_recovery_amount', 'above_threshold']])).fit()

# Display the regression results
print(model_with_threshold.summary())
```

# Regression modeling: adjusting the window

```python
# Filter data for the specified narrower range
narrow_window_data = df[(df['expected_recovery_amount'] >= 950) & (df['expected_recovery_amount'] <= 1050)]

# Add an indicator for being above the threshold
narrow_window_data['above_threshold'] = (narrow_window_data['expected_recovery_amount'] > 1000).astype(int)

# Perform OLS regression
model_narrow_window = sm.OLS(narrow_window_data['actual_recovery_amount'], sm.add_constant(narrow_window_data[['expected_recovery_amount', 'above_threshold']]))
results_narrow_window = model_narrow_window.fit()

# Print regression results
print(results_narrow_window.summary())
```
